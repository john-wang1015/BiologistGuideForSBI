```python
import pickle
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import scipy.io as sio

# Replace 'your_file.pkl' with the path to your .pkl file
file_path = 'syn3rsnl_thetas.pkl'

# Open the file in binary read mode
with open(file_path, 'rb') as file:
    # Load the data from the file
    data = pickle.load(file)
```

```python
sio.savemat("RSNL_posterior_syn3.mat",{'theta':data.values[3]})
```

```python
theta_np_array = data.values[3]

titles = [r'$g_{age}^1$', r'$\tau$', r'$g_{age}^2$']
true_values = np.array([100,16, 300])  
fig, axes = plt.subplots(3, 1, figsize=(8, 12))

for i in range(3):
    sns.kdeplot(theta_np_array[:,i], ax=axes[i], fill=True)
    axes[i].axvline(x=true_values[i], color='k', linewidth=3, label='True Value')
    axes[i].set_title(titles[i])
    axes[i].set_ylabel('Density')
    axes[i].legend()

plt.tight_layout()

# Save the figure
plt.savefig('syn3.png')

# Show the plots
plt.show()
```
